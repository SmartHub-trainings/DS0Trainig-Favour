```python
week_1=np.random.randint(1,10,5)
week_2=np.random.randint(1,10,5)
days=["mon","Tue","Wed","Thu","Fri"]
```

```python
plt.bar(days,week_1)
```

```python
plt.bar(days,week_2)
```

```python
plt.bar(days,week_1)
plt.bar(days,week_2)
```

```python
plt.bar(days,week_1,label="Week 1")
plt.bar(days,week_2,label="Week 2")

plt.legend()
```

## Component bar chart

```python
k_range=np.arange(1,6)
plt.bar(k_range,week_1,label="Week 1",width=0.4)
plt.bar(k_range+0.4,week_2,label="Week 2",width=0.4)
plt.xticks(k_range+0.2,days);
plt.legend()
```

```python
plt.stackplot(days,week_1,week_2,labels=["Week 1","Week 2"])
plt.legend()
plt.savefig("my plot.png")
plt.xlabel("The label")
```
