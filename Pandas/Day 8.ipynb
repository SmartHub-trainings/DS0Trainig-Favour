## min-max absolute scaler

```python
s=pd.Series([-10,20,30,10,-23,-10])
```

```python
s
```

```python
# (x-min)/(max-min)
```

```python
(s.abs()-s.abs().min())/(s.abs().max()-s.abs().min())
```

```python
# min-max scalar
```

```python
(s-s.min())/(s.max()-s.min())
```

```python
## Standard scalar
# (x-mean)/std/
```

```python
(s-s.mean())/s.std()
```

```python
((s-s.mean())/s.std()).mean()
```

```python
((s-s.mean())/s.std()).std()
```
